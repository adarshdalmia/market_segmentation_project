```python
# Importing the necessary classes
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the data from csv file to a Pandas DataFrame
data = pd.read_csv("Indian automoble buying behavour study 1.0.csv")
data.head()
```

```python
# Number of rows and columns in the dataframe
data.shape
```

```python
# Name of all columns in the dataframe
data.columns
```

```python
# Getting more information about the dataset
data.info()
```

```python
# Getting some statistical measures about the column AGE which only has numerical values.
data.describe()
```

```python
# checking for missing values in each column
data.isnull().sum()
```

```python
for column in ['Profession','Marrital Status','Education','Personal loan','House Loan','Wife Working','Make']:
  print(column,':',data[column].unique())
```

```python
# Visualization based on profession
sns.countplot(data=data, x='Profession',palette="Paired")
plt.title('Profession Analysis')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.show()
```

```python

profession_counts = data['Profession'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))  # Set the figure size
colors = ['#FFFF00','#FF00FF']
explode = (0.01,0.01)
my_circle=plt.Circle( (0,0), 0.5, color='white')
plt.pie(profession_counts,colors = colors, labels=profession_counts.index, autopct='%1.1f%%', startangle=90,pctdistance=0.80,explode=explode)
plt.title('Profession Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
# Histogram based on customer age
sns.histplot(data=data, x='Age', bins=10,color='pink', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization based on marital status
plt.figure(figsize=(6, 4))
colors = ['#FFFF00','#FF00FF']
sns.countplot(data=data, x='Marrital Status',palette="dark")
plt.title('Marital Status Analysis')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
from statsmodels.graphics.mosaicplot import mosaic
plt.figure(figsize=(10, 6))
mosaic(data, ['Education'], title='Education Analysis')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.swarmplot(data=data, x='Education', palette="cubehelix")
plt.title('Education Analysis')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization based on educational qualification
plt.figure(figsize=(7, 5))
sns.countplot(data=data, x='Education',palette="cubehelix")
plt.title('Education Analysis')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=data, x='Total Salary', color='blue', fill=True)
plt.title('Total Salary Distribution (Kernel Density)')
plt.xlabel('Total Salary')
plt.ylabel('Density')
plt.show()
```

```python
# Total Salary Distribution visualization
plt.figure(figsize=(5, 3))
sns.histplot(data=data, x='Total Salary', bins=10,color='orange', edgecolor='black')
plt.title('Total Salary Distribution')
plt.xlabel('Total Salary')
plt.ylabel('Count')
plt.show()
```

```python
#Vehicle Make Analysis
plt.figure(figsize=(6, 3))
sns.countplot(data=data, x='Make',palette="dark")
plt.title('Vehicle Make Analysis')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = data.corr()

# Set up the figure and axes
plt.figure(figsize=(8, 6))
plt.title('Correlation Heatmap')

# Create the heatmap with more customization
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, linewidths=0.5)

# Adjust tick labels
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.show()
```

```python
g = sns.PairGrid(data)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
```
